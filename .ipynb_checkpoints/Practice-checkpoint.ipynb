```python
#creating a login page
import tkinter as tk
a=tk.Tk()
a.title("My Login")
a.iconbitmap("a.ico")
a.maxsize(width=500,height=400)
a.configure(background="White")
data1=tk.Label(text="LOGIN",bg="white",fg="Red",font=("times",30,"bold"))
data1.grid(row=0,column=1,padx=100)
data=tk.Label(text="Login Id",fg="Black",bg="white",font=("times",22,'bold'))
data.grid(row=1,column=0,padx=0,pady=50)
entry1=tk.Entry(bg="White",fg="Black",font=("times",12,"bold"))
entry1.grid(row=1,column=1)
password=tk.Label(text="Password",bg="white",fg="black",font=("times",22,"bold"))
password.grid(row=2,column=0)
pass_entry=tk.Entry(bg="white",fg="black",show="*",font=("times",12,"bold"))
pass_entry.grid(row=2,column=1)
buttton=tk.Button(text="Submit",bg="red",fg="black",font=("times",12,"bold"))
buttton.grid(row=7,column=1,padx=40,pady=100)
a.mainloop()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Line graph
data=[1,22,20,4,558,74]
months=["jan",'feb','mar','april','may','june']
plt.xlabel("months")
plt.ylabel("data")
plt.title("Salary")
plt.plot(months,data,color="purple",linestyle="--",marker="*")
plt.show()
```

```python
#bar graph
info=["ankit","rohan",'rajesh','mohit','harun']
age=[23,25,45,47,32]
plt.xlabel("Names")
plt.ylabel("age")
plt.title("Age of the Students")
plt.barh(info,age,color="black")
plt.show()
```

```python
#Scatter plot
info=["ankit","rohan",'rajesh','mohit','harun',"nivu"]
age=[23,25,45,47,32,88]
plt.xlabel("Names")
plt.ylabel("age")
plt.title("Age of the Students")
plt.scatter(info,age,color="black",marker="*")
plt.show()
```

```python
#histogram
a=[10,50,80,59,82,78,153,123,10,48,80,50]
plt.xlabel('Salary')
plt.ylabel("No of People")
plt.title("Salary Graph")
plt.hist(a,bins=4,color="purple",edgecolor="black")
plt.show()
```

```python
#pie char
info=["ankit","rohan",'rajesh','mohit','harun',"nivu"]
sale=[2300,2500,4075,4457,3256,8678]
b=[0.1,0,0,0,0,0.4]
plt.xlabel("Names")
plt.ylabel("sale")
plt.title("sale Students")
plt.pie(sale,labels=info,shadow="True",autopct="%1.1f%%",explode=b)
plt.show()
```

```python
#stack plot 
name=["ankit","rohan",'rajesh','mohit','harun',"nivu"]
age=[232,251,454,478,325,88]
salary=[2300,2500,4075,4457,3256,8678]
saving=[100,200,300,400,500,600]
plt.stackplot(name,age,salary,saving,labels=(age,salary,saving))
plt.show()
```
